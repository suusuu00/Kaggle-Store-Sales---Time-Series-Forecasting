```python
# LSTM 전용 CustomDataset
class CustomDataset(Dataset):
    def __init__(self, dataX, dataY, sequence_length):
        self.dataX = dataX
        self.dataY = dataY
        self.sequence_length = sequence_length

    def __len__(self):
        return len(self.dataY) - self.sequence_length

    def __getitem__(self, idx):
        inputs = torch.tensor(self.dataX[idx:idx + self.sequence_length].values, dtype=torch.float32)
        label = torch.tensor(self.dataY[idx + self.sequence_length], dtype=torch.float32)
            
        return inputs, label
```

## Model Class 생성 & 선언

```python
class LinearModel(nn.Module):
    def __init__(self, input_features, out_features, hidden_dim_1, hidden_dim_2):
        super().__init__()

        self.layer = nn.Sequential(
            nn.Linear(in_features= input_features, out_features=hidden_dim_1),
            nn.ReLU(),
            nn.Linear(in_features= hidden_dim_1, out_features= hidden_dim_2),
            nn.ReLU(),
            nn.Linear(in_features= hidden_dim_2, out_features= out_features)
        )
        
    def forward(self, x):
        x = self.layer(x)
        
        return x
        
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = LinearModel(input_features= merge_train_oe_df.shape[1], out_features= 1, hidden_dim_1= 128, hidden_dim_2= 64).to(device)
```

```python
class LSTMModel(nn.Module):
    def __init__(self, input_size, output_size, hidden_dim_1, hidden_dim_2, num_layers = 1):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_dim_1, num_layers, batch_first=True)
        self.fc = nn.Sequential(
            nn.Linear(hidden_dim_1, hidden_dim_2),
            nn.ReLU(),
            nn.Linear(hidden_dim_2, output_size)
        )

    def forward(self, x):
        out, _ = self.lstm(x)
        out = self.fc(out[:, -1])
        return out
    
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = LSTMModel(input_size= merge_train_oe_df.shape[1], output_size= 1, hidden_dim_1=128, hidden_dim_2=64)
```

## Optimizer

```python
### https://github.com/Bjarten/early-stopping-pytorch/blob/master/pytorchtools.py
class EarlyStopping:
    """Early stops the training if validation loss doesn't improve after a given patience."""
    def __init__(self, patience=7, verbose=False, delta=0, path='checkpoint.pt', trace_func=print):
        """
        Args:
            patience (int): How long to wait after last time validation loss improved.
                            Default: 7
            verbose (bool): If True, prints a message for each validation loss improvement.
                            Default: False
            delta (float): Minimum change in the monitored quantity to qualify as an improvement.
                            Default: 0
            path (str): Path for the checkpoint to be saved to.
                            Default: 'checkpoint.pt'
            trace_func (function): trace print function.
                            Default: print
        """
        self.patience = patience
        self.verbose = verbose
        self.counter = 0
        self.best_score = None
        self.early_stop = False
        self.val_loss_min = np.Inf
        self.delta = delta
        self.path = path
        self.trace_func = trace_func

    def __call__(self, val_loss, model): ## __call__ 함수는 instance 객체를 생성할때 자동으로 class의 객체를 생성하도록 하는 함수 // call함수의 입력인자값은 earlystopping 객체를 실행할때 입력

        score = -val_loss

        if self.best_score is None:     # 처음에 best_score에 값이 없을 때, 초기 score값 지정
            self.best_score = score
            self.save_checkpoint(val_loss, model)
        elif score < self.best_score + self.delta:    # score가 오르지 않았을 때
            self.counter += 1
            self.trace_func(f'EarlyStopping counter: {self.counter} out of {self.patience}')
            if self.counter >= self.patience:
                self.early_stop = True
        else:                           # score가 잘 오르고 있을 때
            self.best_score = score
            self.save_checkpoint(val_loss, model)
            self.counter = 0

    def save_checkpoint(self, val_loss, model):
        '''Saves model when validation loss decrease.'''
        if self.verbose:
            self.trace_func(f'Validation loss decreased ({self.val_loss_min:.6f} --> {val_loss:.6f}).  Saving model ...')
        torch.save(model.state_dict(), self.path)
        self.val_loss_min = val_loss

os.makedirs('./models', exist_ok = True)
early_stopping = EarlyStopping(patience = 5, verbose = True, path = './models/linear_model4.pt' )
```

RMSLE = $\sqrt{ \frac{1}{n} \sum_{i=1}^n \left(\log (1 + \hat{y}_i) - \log (1 + y_i)\right)^2}$

```python
class RMSLELoss(nn.Module):
    def __init__(self):
        super().__init__()
        self.mse = nn.MSELoss()
        
    def forward(self, pred, actual):
        y_true_log = torch.log1p(actual)
        y_pred_log = torch.log1p(pred)
        # 매우 작은 값에 대한 log 연산이 NaN loss를 유발
        # NaN 값 제거
        y_true_log = torch.where(torch.isfinite(y_true_log), y_true_log, torch.zeros_like(y_true_log))
        y_pred_log = torch.where(torch.isfinite(y_pred_log), y_pred_log, torch.zeros_like(y_pred_log))
        
        return torch.sqrt(self.mse(y_pred_log, y_true_log))
```

```python
optimizer = optim.Adam(params = model.parameters(), lr = 1e-3)
criterion = RMSLELoss().to(device)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience = 3, factor = 0.1, verbose = True) ## factor : lr 감소시킬 비율
```

## Train

### Linear

#### 이전에 훈련한 linear

date 열 year, month, day로 안 바꾸고 돌리고 전체적으로 pca하지는 않은 linear model
- val RMSLE : 0.218691
- submit한 RMSLE : 1.07299

```python
epochs = 100

for epoch in range(epochs):
    
    train_cost = 0
    val_cost = 0
    
    model.train()
    torch.autograd.set_detect_anomaly(True)
    for train_inputs, train_labels in train_dataloader:
        train_inputs = train_inputs.to(device)
        train_labels = train_labels.to(device)
        
        train_preds = model(train_inputs)
        
        train_loss = criterion(train_preds.flatten(), train_labels)
        
        optimizer.zero_grad()
        train_loss.backward()
        optimizer.step()
        
        train_cost += train_loss.item()
        
    train_cost /= len(train_dataloader)
    
    model.eval()
    with torch.no_grad():
        for val_inputs, val_labels in val_dataloader:
            val_inputs = val_inputs.to(device)
            val_labels = val_labels.to(device)
            
            val_preds = model(val_inputs)
            
            val_loss = criterion(val_preds.flatten(), val_labels)
            
            val_cost += val_loss.item()
            
    val_cost /= len(val_dataloader)
    
    scheduler.step(val_cost)

    early_stopping(val_cost, model)
    if early_stopping.early_stop:
        break
    
    print(f"{epoch + 1} / {epochs} - train_cost : {train_cost}, val_cost : {val_cost}")
```
