# 원 핫 인코딩

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.decomposition import PCA
```

```python
# 카테고리 컬럼
category_col = ['city', 'state', 'store_type', 'holiday_type', 'locale', 'locale_name', 'transferred']
# 범주가 너무 많은 컬럼
many_value_col = ['family', 'description']
```

```python
many_col_oe = OneHotEncoder(sparse_output = False, handle_unknown = 'infrequent_if_exist')
many_col_oe.fit(merge_train_df[[many_value_col[0]]])
train_family_df = pd.DataFrame(many_col_oe.transform(merge_train_df[[many_value_col[0]]]), columns = many_col_oe.get_feature_names_out())
test_family_df = pd.DataFrame(many_col_oe.transform(merge_test_df[[many_value_col[0]]]), columns = many_col_oe.get_feature_names_out())
```

```python
train_family_df.shape
```

```python
family_pca = PCA(n_components=26)
family_pca.fit(train_family_df)
family_select_col = family_pca.transform(train_family_df)
family_pca.explained_variance_ratio_.sum()
```

```python
many_col_oe = OneHotEncoder(sparse_output = False, handle_unknown = 'infrequent_if_exist')
many_col_oe.fit(merge_train_df[[many_value_col[1]]])
train_description_df = pd.DataFrame(many_col_oe.transform(merge_train_df[[many_value_col[1]]]), columns = many_col_oe.get_feature_names_out())
print(train_description_df.shape)

description_pca = PCA(n_components=40)
description_pca.fit(train_description_df)
description_select_col = description_pca.transform(train_description_df)
description_pca.explained_variance_ratio_.sum()
```

```python
oe = OneHotEncoder(sparse_output = False, handle_unknown = 'infrequent_if_exist')
oe.fit(merge_train_df[category_col])
train_oe_df = pd.DataFrame(oe.transform(merge_train_df[category_col]), columns = oe.get_feature_names_out())
```

```python
merge_train_oe_df = pd.concat([merge_train_df.reset_index(drop = True, inplace = False), family_df, description_df, oe_df], axis = 1)
merge_train_oe_df.drop(columns=category_col + many_value_col)
```
